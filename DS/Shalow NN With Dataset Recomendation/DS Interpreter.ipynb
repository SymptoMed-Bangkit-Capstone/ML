# Inference Model

```python
model = tf.keras.models.load_model('save_model_nn')
interpreter = tf.lite.Interpreter("model_nn.tflite")
interpreter.allocate_tensors()

# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
# Test the TensorFlow Lite model on random input data.
data_x = df_custom.to_numpy()
data_y = df_custom_true.to_numpy()
# np.array(np.random.random_sample(input_shape), dtype=np.float32)
# input_shape = input_details[0]['shape']
inputs, outputs, labels = [], [], []
input_data = np.array(data_x[0].reshape(1, data_x[0].shape[0]), dtype=np.float32)
# print(input_data)
interpreter.set_tensor(input_details[0]['index'], input_data)

interpreter.invoke()
tflite_results = interpreter.get_tensor(output_details[0]['index'])

# Test the TensorFlow model on random input data.
constant_data = tf.constant(input_data)
# print(constant_data)
tf_results = model(constant_data)
output_data = np.array(tf_results)

inputs.append(input_data[0][0])
outputs.append(output_data[0][0])
labels.append(output_data)
```

```python
data_rekomendasi = pd.read_csv("./data_rekomendasi.csv", sep=';')
```

```python
index_predict = np.argmax(labels[0])

indeks_hasil = data_rekomendasi[data_rekomendasi['Index'] == index_predict]
probability = f"{round((labels[0][0][index_predict])*100, 2)}%"
symptom = indeks_hasil['Symptom'].values[0]
link = indeks_hasil['Detail'].values[0]
saran = indeks_hasil['Saran'].values[0]
truesymptom = data_rekomendasi[data_rekomendasi['Index'] == data_y[0][0]]['Symptom'].values[0]


print(f"Kelas : {symptom}")
print(f"True Kelas: {truesymptom}")
print(f"Probabilitas : {probability}")
print(f"link : {link}")
print(f"Rekomendasi : {saran}")
```
